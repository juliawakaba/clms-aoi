## 3. Run DynamicLandCover for an AOI

```python
from clms_aoi.aoi import BoundingBox
from clms_aoi.products.dynamic_land_cover import DynamicLandCover

# Small test area over central Portugal
bbox = BoundingBox(west=-8.5, south=39.5, east=-8.0, north=40.0)
geometry = {
    "type": "Polygon",
    "coordinates": [[
        [-8.5, 39.5],
        [-8.0, 39.5],
        [-8.0, 40.0],
        [-8.5, 40.0],
        [-8.5, 39.5],
    ]]
}

product = DynamicLandCover(sh, base_url=sh.sh_base_url)
df = product.visualize(bbox, geometry, year=2024)
df
```
